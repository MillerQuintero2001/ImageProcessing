## Actividad 3

En la actividad 1 se aprendi칩 a manipular la matriz de la imagen para dividirla en cuadrantes, la segunda actividad fue de transformar la imagen en diferentes espacios de color y c칩mo se debe de dividir. En esta actividad se le pide que desarrolle manualmente un algoritmo que le permita transformar una imagen RGB a YUV. La transformaci칩n de RGB a YUV se realiza mediante la siguiente f칩rmula: 

```python
Image("res/actividad3/yuv.png")
```

```python
img = cv2.imread("res/actividad2/tigre.png", cv2.IMREAD_COLOR)
# Img rgb
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_yuv = img_rgb.copy()
# Calcular YUV
img_yuv[:,:,0] = 0.299*img_rgb[:,:,0] + 0.587*img_rgb[:,:,1] + 0.114*img_rgb[:,:,2]
img_yuv[:,:,1] = -0.147*img_rgb[:,:,0] + 0.289*img_rgb[:,:,1] + 0.436*img_rgb[:,:,2]
img_yuv[:,:,2] = 0.615*img_rgb[:,:,0] - 0.515*img_rgb[:,:,1] - 0.100*img_rgb[:,:,2]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 10))
# Y
axs[0].imshow(img_yuv[:,:,0], cmap="gray")
axs[0].set_title("Y")
# U
axs[1].imshow(img_yuv[:,:,1], cmap="gray")
axs[1].set_title("U")
# V
axs[2].imshow(img_yuv[:,:,2], cmap="gray")
axs[2].set_title("V")
```
